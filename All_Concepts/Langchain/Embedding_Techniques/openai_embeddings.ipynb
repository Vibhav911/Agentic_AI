```python
from langchain_community.document_loaders import TextLoader
docs = TextLoader("/home/vibhav911/Documents/skills/Data Science/Langchain/Data_Ingestion/speech.txt").load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitters = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 50)
final_document = text_splitters.split_documents(docs)
final_document
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
# Vector Embeddings and Vector Databases
from langchain.vectorstores import Chroma

db = Chroma.from_documents(final_document, embeddings)
db
```

```python
text = "easier for us to conduct ourselves as belligerents in a high spirit"
retrieved_results = db.similarity_search(text)
print(retrieved_results)
```
