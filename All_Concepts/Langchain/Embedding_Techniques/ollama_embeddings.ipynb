# Ollama Embeddings

##### Ollama supports embedding models, making it possible to build retrieval augmented generation (RAG) applications that combine text prompts with existing documents or other data.

```python
from langchain_ollama import OllamaEmbeddings

embeddings = OllamaEmbeddings(model = "gemma2:2b")  # By default, it uses llama model
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is the first letter of the greek alphabets",
        "Beta is the second letter of the greek alphabets"
    ]
)
```

```python
len(r1[1])
```

```python
embeddings.embed_query("What is the second letter of the greek alphabets")
```
